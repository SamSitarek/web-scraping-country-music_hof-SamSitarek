Question 4: Now, you need to take what you created for the first page and apply it across the rest of the pages so that you can scrape all inductees. Notice that when you click the next page button at the bottom of the page that the url changes to "https://countrymusichalloffame.org/hall-of-fame/members/page/2". Check that the code that you wrote for the first page still works for page 2.
```python
artist_all=[]
year_all=[]

for i in range (1,17):
        website_url_all = 'https://countrymusichalloffame.org/hall-of-fame/members/page/'+str(i)
        print(website_url_all)
        response_all = requests.get(website_url_all)
        print(response_all.status_code)
        print(type(response_all))
        response_all.content
        soup_all = BeautifulSoup(response_all.content, 'lxml')
        print(soup_all.title)
        print(soup_all.prettify())
        for name in soup_all.find_all('h3'):
            print(name.string)
            artist_all.append(name.string)
        for year in soup_all.find_all('div', attrs = {'class': 'vertical-card_content--copy'}):
            year = str(year.text)
            year = year.strip('\n\n                    Inducted ')
            year_all.append(year)
```
